# 2.0. Feature Engineering

```python
df_feature = df_description.copy()
```

```python
df_feature['gross_revenue'] = df_feature.quantity * df_feature.unit_price

# Recency
diff = df_feature.invoice_date.max() - df_feature.groupby('customer_id').max()['invoice_date']
recency = pd.DataFrame(diff).reset_index().rename(columns={'invoice_date': 'recency'})

recency['recency'] = recency['recency'].dt.days
df_feature = df_feature.merge((recency), how='left', on='customer_id')


# Frequency
frequency = df_feature.groupby('customer_id').nunique()[['invoice_no']].reset_index().rename(columns={'invoice_no': 'frequency'})
df_feature = df_feature.merge((frequency), how='left', on='customer_id')

# Monetary
monetary = df_feature.groupby('customer_id').sum(numeric_only=True).gross_revenue.reset_index().rename(columns={'gross_revenue': 'monetary'})
df_feature = df_feature.merge((monetary), how='left', on='customer_id')
```

# 7.0. Hyperparameter Fine-Tuning

DBSCAN -> O dbscan identifica pontos que estão aglomerados. Temos que definir dois parâmetro o mínimo de amostras e a distância eps. Se houver o mínimo de amostra a uma distância eps de um ponto esse ponto será um "centro de cluster" Se tivermos "centro de clusters" com uma distância entre eles menor do que eps, eles são colocados no mesmo grupo.

1. Escolhe pontos de forma arbitrária.
2. Encontra todos os ponto a uma distância eps
3. O ponto inicial será classificado com "ruído" se não houver pontos suficientes a uma distânci eps (mínimo de amostra).
4. Os vizinhos são visitados e se eles não pertencerem a nenhum cluster ainda, eles serão classificado com o novo cluster.
5. O cluster cresce até acabar os centros com o mínimo de amostra a uma distânci eps.

## 7.2 Sillhoutte Score

```python
import numpy as np
```

```python
ss_list = []
eps_list = np.arange(55, 300, 5)
for k in eps_list:


    # model
    dbscan = DBSCAN(eps=k, min_samples=5)

    # model training
    labels = dbscan.fit_predict(X)
    ss = mt.silhouette_score(X, labels)
    ss_list.append(ss)

plt.figure(figsize=(10, 6))
plt.plot(eps_list, ss_list, linestyle='--', marker='o', color='b')
plt.xlabel('eps')
plt.ylabel('Silhouette Score')
plt.title('SS vs eps');
```
